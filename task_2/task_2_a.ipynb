Let see at the result modul of creating and visualization of graphs

```python
import graph_function
```

First we should create object of graph

```python
graph_1 = graph_function.Graph()
```

Then add some nodes with info. (We don't set any coordinates since module set it to a vertices of a regular polygon)

```python
graph_1.add_node("A", "Node A")
graph_1.add_node("B", "Node B")
graph_1.add_node("C", "Node C")
graph_1.add_node("D", "Node D")
graph_1.add_node("E", "Node E")
```

And some edges

```python
graph_1.add_edge("A", "B")
graph_1.add_edge("B", "C")
graph_1.add_edge("E", "B")
```

Now let's print our graph

```python
graph_1.draw_graph()
```
