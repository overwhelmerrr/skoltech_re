Now let's try to choose friends for best party ever

```python
import additional_functions
import graph_function
import bron_kerbosh
import random
```

First let's try it on specific graph

```python
graph_1 = graph_function.Graph()
```

```python
graph_1.add_node("0", "Putin")
graph_1.add_node("1", "Lukashenko")
graph_1.add_node("2", "Afewerki")
graph_1.add_node("3", "Mr Xi")
graph_1.add_node("5", "Scholz")
graph_1.add_node("6", "Macron")
graph_1.add_node("4", "Erdogan")
graph_1.add_node("7", "Zelensky")
graph_1.add_node("8", "Biden")
```

```python
graph_1.pictures = True
```

```python
graph_1.add_edge("0", "5")
graph_1.add_edge("0", "6")
graph_1.add_edge("0", "7")
graph_1.add_edge("0", "8")
graph_1.add_edge("1", "7")
graph_1.add_edge("2", "7")
```

```python
graph_1.draw_graph()
```

Now we create complement_graph and find maximum cliques of it

```python
adjacency_dict = graph_1.create_adjacency_dict()

complement_graph = additional_functions.build_complement_graph(adjacency_dict)

result = bron_kerbosh.bron_kerbosch_max_by_inclusion(complement_graph)
```

```python
print(f"Maximum we can invite {len(result[0])} person.")
print('For example our best party ever members are:')
for clique in result:
    print("Nodes with id",*clique)
    graph_1.draw_graph(show_clique=clique)
    print()
```

```python
nodes = random.randint(3, 8)
edges = random.randint(0, nodes*(nodes-1)/2)

graph_2 = graph_function.Graph()
graph_2.generate_random_graph(nodes, edges)
graph_2.pictures = True
graph_2.draw_graph()
```
